**1) Program for Calculating the Area of a Circle**

```python
# Develop a python prpogram for determining the area of a circle
# The equation for the area of a circle requires a stipulated radius and a constant pi
# the constahnt pi has to be imported from the mtah package of python
import math 
#print the value of pi
print( 'The value of pi for this program is =' , str(math.pi))
pi=math.pi 
# input the measurement of radius
r=int(input('Enter the value of radius = '))
#developing the equation of area
area = pi*r**2
#print the final value
print('The area of the circle is =' , str(area), 'square units')

```

After providng the value of radious as 6 units and importing the value of pi the area of the circle  is calculated. The unit of the measurement can be set up according to the requirement of the user.

**3) Program for caluclating the Area of a Sector of a Circle**

```python
# Develop a python prpogram for determining the area of sector of a circle
# The equation for the area of a circle requires a stipulated radius, central angle and a constant pi
# the constahnt pi has to be imported from the mtah package of python
import math 
#print the value of pi
print( 'The value of pi for this program is =' , str(math.pi))
pi=math.pi 
# input the measurement of radius
r=int(input('Enter the value of radius = '))
# input the value of central angle
a=int(input('Enter the measure of central angle = ')) 
#developing the equation of area of a sector
area = (a/360)*(pi*r**2)
# for limtng the number of digts after decimal point str.format(*args) function is used
limited_float = "{:.2f}".format(area)
#print the final value
print('The area of sector of the circle is =' , str(limited_float), 'square units')
```

After providng the value of radious as 10 units and providng the value of the central angle as 90 degrees the area of sector of the circle is calculated. The unit of the measurement can be set up according to the requirement of the user.
